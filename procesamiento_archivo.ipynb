```python
import pandas as pd
import re
```

```python
FILE = 'Final_Data_Hackathon_2022.csv'
data = pd.read_csv(FILE)
```

```python
fills = [
    'Nombre1',
    'Nombre2',
    'Nombre3',
    'Direccion1',
    'Direccion2',
    'Direccion3'
]

for f in fills:
    data[f] = data[f].fillna('')
```

```python
data['Nombre1'] = data['Nombre1'].astype(str)
data['Nombre2'] = data['Nombre2'].astype(str)
data['Nombre3'] = data['Nombre3'].astype(str)
```

```python
def nombre_completo(row):
    return row['Nombre1'] + row['Nombre2'] + row['Nombre3']
```

```python
data['nombre_completo'] = data.apply(nombre_completo, axis=1)
```

```python
data[0:40]
```

```python
def limpieza(texto):
    patrones = [
        "(SA DE CV)",
        "(SA CV)",
        "(S.A. DE CV)",
        "(S.A DE CV)",
        "(S.A.DE CV)",
        "(SADECV)",
        "(SADE CV)",
        "(SA DECV)",
        "(S DE RL DE CV)",
        "(S DERL DE CV)",
        "(SRL DE CV)",
        "(SDE RL DE CV)",
        "(SC DE RL DE CV)",
        "(S DERL DE CV)",
        "(SPR DE RL DE CV)",
        "(SPR DERL DE CV)"
    ]
    
    n1 = texto.replace("#", "")
    n2 = n1.replace(",", "")
    
    pat_ = "|".join(patrones)
    n3 = re.sub(pat_, "", n2)
    
    return n3
```

```python
data['nombre_completo_limpio'] = data['nombre_completo'].map(limpieza)
```
