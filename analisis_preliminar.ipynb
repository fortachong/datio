```python
import pandas as pd
import re
import numpy as np
```

```python
FILE = 'datio_scraper/data/info/info_0.csv'
```

```python
data = pd.read_csv(FILE, sep='|')
```

```python
data[0:1].T
```

```python
cols = ['id_establecimiento', 'clee', 'nom_estab', 'raz_social', 'entidad', 'municipio', 'cod_postal', 'codigo_act', 
 'nombre_act', 'fecha_alta', 'www', 'latitud', 'longitud', 'per_ocu']
```

```python
d0 = data[cols]
```

```python
d0['per_ocu'].unique()
```

```python
def min_max_ocupacion(st):
    minimo = np.NINF
    maximo = np.PINF
    m = re.findall("[0-9]+", st)
    if len(m) == 1: 
        minimo = m[0]
    elif len(m) == 2: 
        minimo = m[0]
        maximo = m[1]
    return minimo, maximo
```

```python
min_max_ocupacion('251 y más personas')
```

```python
d0['min_ocupacion'] = d0['per_ocu'].map(lambda x: min_max_ocupacion(x)[0])
d0['max_ocupacion'] = d0['per_ocu'].map(lambda x: min_max_ocupacion(x)[1])
```
